# Function to Search all files

```python
def SearchAboutFiles (path, fileType):
    global FilesNames
    FilesNames = []
    for root, dirs, files in os.walk(path, topdown=False):
        for name in files:
            if name.endswith(fileType):
                FileNameAndPath = []
                root = root.replace('\\','/')
                FileNameAndPath.append(root)
                FileNameAndPath.append(name)
                #print(os.path.join(root, name))
                FilesNames.append(FileNameAndPath)
                
        for value in FilesNames[:]:
            if value[1] == 'classes.txt':
                FilesNames.remove(value)
    return FilesNames
```
